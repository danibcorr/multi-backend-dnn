```python
from illia.nn import Conv2d, Linear
from illia.distributions import GaussianDistribution
```

```python
print(f"Actual backend: {illia.BackendManager.get_backend()}")
print(f"Available backends and modules: {illia.BackendManager._backend_modules}")
```

```python
# --- Linear ---
linear = Linear(in_features=10, out_features=5)
print(f"Linear: {linear}")

# --- Conv2d ---
conv = Conv2d(in_channels=3, out_channels=16, kernel_size=3)
print(f"Conv2d: {conv}")

# --- Distribution ---
dist = GaussianDistribution(shape=(10,))
sample = dist.sample()
log_p = dist.log_prob(sample)
print(f"Sample shape: {sample.shape}")
print(f"Log prob:     {log_p}")
print(f"Num params:   {dist.num_params}")
```
